```
import pandas as pd
from bokeh.models import ColumnDataSource
from bokeh.plotting import Figure
from bokeh.palettes import Category20
from bokeh.io import output_notebook, show


output_notebook()
df = pd.read_csv('https://github.com/dkapitan/scik/blob/master/data/aantal-registraties.csv?raw=true', parse_dates=['registratie datum'])
```

```
cds = ColumnDataSource(df)
p = Figure(x_axis_type='datetime', title='Registraties coronainkaart.nl  (lancering t/m 31 maart 2020)')
width = 36*10**6
colors = Category20[3]
stack = ['aantal verdenkingen','aantal bevestigingen']
p.vbar_stack(stack,
             x='registratie datum',
             width=width,
             fill_color=colors[1:],
             line_color=colors[1:],
             legend_label=stack,
             source=cds)
p.line(x='registratie datum', y='aantal huisgenoten', line_color=colors[0], legend_label='aantal huisgenoten', source=cds)
p.legend.location = 'top_left'
p.ygrid.minor_grid_line_color = 'grey'
p.ygrid.minor_grid_line_alpha = 0.1
show(p)
```
